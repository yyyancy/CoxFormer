```python
import os
import pandas as pd
os.chdir(os.path.abspath(".."))
from utils.Cell_level_task_utils import process_dataset_benchmark

datasets_info = [    
("BC4_FLEX", "cell_type"),    
("DLBCL", "cell_type"),    
("Lung", "cell_type"),    
("bonemarrow", "clusters"),    
]

methods = ["SCGPT-HUMAN","SCGPT-PANCANCER","GF-6L30M","GF-20L95M","GF-12L95M","GF-12L30M",
           "GF-12L95MCANCER","GENE2VEC","TCGA-EMBEDDING","FROGS-ARCHS4","MUT2VEC",
           "GENEPT-MODEL3","GENEPT-ADA","BIOCONCEPTVEC-SKIP-GRAM","BIOCONCEPTVEC-FASTTEXT",
           "BIOCONCEPTVEC-GLOVE","BIOCONCEPTVEC-CBOW","CoxFormer"]

file_paths = {m: f"Embeddings/{m}.pkl" for m in methods}

out_dir = "Result/Cell_level_task/"
os.makedirs(out_dir, exist_ok=True)

def run_celltype_annotation(dataset_name, label_key):
    save_file = os.path.join(out_dir, f"{dataset_name}_results.csv")

    if os.path.exists(save_file):
        tmp_df = pd.read_csv(save_file)
        existing_methods = set(tmp_df["Method"].astype(str))
    else:
        tmp_df = None
        existing_methods = set()

    additional_tasks = []
    for method in methods:
        if method not in existing_methods:
            additional_tasks.append((dataset_name, label_key, method))

    if not additional_tasks:
        print(f"[{dataset_name}] all methods already evaluated, skip.")
        return

    print(f"[{dataset_name}] running tasks: {additional_tasks}")

    for (dname, lkey, m) in additional_tasks:
        res = process_dataset_benchmark(dname, lkey, m, file_paths)
        df = pd.DataFrame(res)
        write_header = not os.path.exists(save_file)
        df.to_csv(save_file, mode="a", header=write_header, index=False)
        print(f"[{dataset_name}] appended method={m} -> {save_file}")

def main():
    for dataset_name, label_key in datasets_info:
        run_celltype_annotation(dataset_name, label_key)

if __name__ == "__main__":
    main()
```
