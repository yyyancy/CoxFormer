# Configs

```python
# ======== Basic utilities ========
import os
import warnings

# ======== Scientific computing ========
import numpy as np
import pandas as pd

# ======== Bioinformatics ========
import anndata as ad

# ======== Warning settings ========
warnings.filterwarnings('ignore')
os.chdir(os.path.abspath(".."))
from utils.Imputation_baseline_utils import SpaGE_impute, gimVI_impute, novoSpaRc_impute, SpaOTsc_impute, stPlus_impute, Tangram_impute
```

# Predicting undetected transcripts by All method

```python
datasets = ["CRC1","CRC2","LM1","LM2"]
for dataset_idx in datasets:
    DataDir = f'Dataset/Pathological_region_detection/{dataset_idx}/'
    outdir = f'Result/Pathological_region_detection/{dataset_idx}/'
    RNA_path = DataDir + 'scrna.tsv'
    Spatial_path = DataDir + 'cnts.tsv'
    location_path = DataDir + 'locs.tsv'
    
    # load RNA data
    RNA_data = pd.read_table(RNA_path, sep='\t', header=0, index_col=0)
    RNA_adata = ad.AnnData(RNA_data)
    
    # load spatail data
    Spatial_data = pd.read_table(Spatial_path, sep='\t', header=0, index_col=0)
    Spatial_adata = ad.AnnData(Spatial_data)
    
    # load location data
    locs_df = pd.read_table(location_path, header=0, index_col=0)
    locations = locs_df[["x","y"]].values
    
    # load train/predict data
    train_gene_full = np.load(DataDir + 'genes_train.npy', allow_pickle = True).tolist()
    predict_gene_full = np.load(DataDir + 'genes_test.npy', allow_pickle = True).tolist()
    
    train_gene = list(set(train_gene_full) & set(Spatial_data.columns))
    predict_gene = list(set(predict_gene_full) & set(Spatial_data.columns))

    # merge gene lists
    selected_genes = list(train_gene + predict_gene)
    print(f"Total selected genes: {len(selected_genes)}")
    
    # subset RNA data
    RNA_data = RNA_data[selected_genes].copy()
    RNA_adata = RNA_adata[:, selected_genes].copy()
    
    # subset Spatial data
    Spatial_data = Spatial_data[selected_genes].copy()
    Spatial_adata = Spatial_adata[:, selected_genes].copy()

    device = 'GPU'
    if not os.path.exists(outdir):
        os.mkdir(outdir)
    
    # ===== stPlus =====
    stplus_path = os.path.join(outdir, "stPlus_impute.csv")
    if not os.path.exists(stplus_path):
        stPlus_result = stPlus_impute()
        stPlus_result.to_csv(stplus_path, index=False)
    else:
        print(f"Skip stPlus: {stplus_path} already exists.")
        
    # ===== gimVI =====
    gimvi_path = os.path.join(outdir, "gimVI_impute.csv")
    if not os.path.exists(gimvi_path):
        gimVI_result = gimVI_impute()
        gimVI_result.to_csv(gimvi_path, index=False)
    else:
        print(f"Skip gimVI: {gimvi_path} already exists.")
    
    # ===== SpaGE =====
    spage_path = os.path.join(outdir, "SpaGE_impute.csv")
    if not os.path.exists(spage_path):
        SpaGE_result = SpaGE_impute()
        SpaGE_result.to_csv(spage_path, index=False)
    else:
        print(f"Skip SpaGE: {spage_path} already exists.")
    
    # ===== novoSpaRc =====
    novosparc_path = os.path.join(outdir, "novoSpaRc_impute.csv")
    if not os.path.exists(novosparc_path):
        novoSpaRc_result = novoSpaRc_impute()
        novoSpaRc_result.to_csv(novosparc_path, index=False)
    else:
        print(f"Skip novoSpaRc: {novosparc_path} already exists.")   
    
    # ===== Tangram =====
    tangram_path = os.path.join(outdir, "Tangram_impute.csv")
    if not os.path.exists(tangram_path):
        Tangram_result = Tangram_impute()
        Tangram_result.to_csv(tangram_path, index=False)
    else:
        print(f"Skip Tangram: {tangram_path} already exists.")
        
    # ===== SpaOTsc =====
    spaotsc_path = os.path.join(outdir, "SpaOTsc_impute.csv")
    if not os.path.exists(spaotsc_path):
        SpaOTsc_result = SpaOTsc_impute()
        SpaOTsc_result.to_csv(spaotsc_path, index=False)
    else:
        print(f"Skip SpaOTsc: {spaotsc_path} already exists.")
```
